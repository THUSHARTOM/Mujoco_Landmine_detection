```python
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
import pandas as pd
```

```python
points = pd.read_csv('file_test.csv')
```

```python
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

x = points['X_Axis'].values
y = points['Y_Axis'].values
z = points['depth_mm'].values * -1
value = points['pressure_N']
scatter = ax.scatter(x, y, z, c=value, s=20, cmap='hot')
fig.colorbar(scatter, ax=ax)

plt.show()
```
